```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# A custom function to calculate
# probability distribution function
def pdf(x, mean, std):
    y_out = 1 / (std * np.sqrt(2 * np.pi)) * np.exp(-((x - mean) ** 2) / (2 * std ** 2))
    return y_out

# Define parameters for the normal distribution
mean = 0
std = 1

# Generate an array of x-values
x = np.arange(-4, 4, 0.1)

# Generate an array of y-values using the PDF
y = pdf(x, mean, std)

# Plotting the bell-shaped curve
#plt.style.use('seaborn')
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='black', linestyle='dashed', label='PDF')
plt.scatter(x, y, marker='o', s=25, color='red', label='Data Points')
plt.title("Probability Density Function (Normal Distribution)")
plt.xlabel("x")
plt.ylabel("PDF(x)")
plt.legend()
plt.grid(True)
plt.savefig('bell_curve.png')
plt.show()
```
